```python
import os

sentences = []

def get_downloaded_tweets(path):
    files = os.listdir(path)
    for sentence_batch in files:
        with open(path + '/' +sentence_batch, 'rb') as f:
            for line in f:
                sentences.append(line.split(' '))
    print len(sentences)

get_downloaded_tweets('spanish_billion_words')
```

```python
# import modules & set up logging
import gensim

def primer((a,b)):
    return a

 # train word2vec on the two sentences
print('creating model')
model = gensim.models.Word2Vec(size=50000, window=3, min_count=3, workers=12)
print 'adding more words'
model.build_vocab(sentences[len(sentences)/2:])
model.train(sentences[:len(sentences)/2])
print('Finish training')
model.init_sims(replace=True)
model.save('/tmp/vectors.txt')
```
