```python
# coding=utf-8
# import modules & set up logging

import gensim
import time
import os, sys

from gensim import models
from gensim.models.word2vec import logger, FAST_VERSION
import logging

logger.setLevel(logging.DEBUG)
print FAST_VERSION
ch = logging.StreamHandler(sys.stdout)
ch.setLevel(logging.INFO)
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
ch.setFormatter(formatter)
logger.addHandler(ch)


sentences = []

def get_downloaded_tweets(path):
    files = os.listdir(path)
    for sentence_batch in files:
        print(sentence_batch)
        with open(path + '/' +sentence_batch, 'rb') as f:
            for line in f:
                sentences.append(line)

get_downloaded_tweets('spanish_billion_words')


def primer((a,b)):
    return a

    
def main():
    # train word2vec on the two sentences
    print 'Starting training'
    model = gensim.models.Word2Vec(sentences, size=100000000, window=3, min_count=5, workers=8,sg=1)
    print 'End of  training'
    model.init_sims(replace=True)
    emojis= ["FACE_WITH_TEARS_OF_JOY",'ley','feliz', 'triste', 'alegre', 'enamorado']
    for e in emojis:
        print(model.most_similar(positive=[e], topn=5))
        lista=[]
        for i in range(5):
            lista.append(primer((model.most_similar(positive=[e], negative=lista, topn=1)).pop()))
        print(e,lista)
        

if __name__ == "__main__":
    main()
```
